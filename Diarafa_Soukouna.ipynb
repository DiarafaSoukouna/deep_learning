(a): 

-La variable thalach ("fréquence cardiaque maximale atteinte") est une variable quantitative discrète.

Explication: La fréquence cardiaque se mesure à travers des valeurs numériques, ce qui en fait une variable quantitative discrète. Explication: La variable age dans le contexte de notre problème est considérée comme une variable quantitative discrète car elle consiste en des valeurs entières et non présentées sous forme d'intervalles.

-La variable chol ("quantité de cholestérol") est une variable quantitative discrète car elle consiste en des valeurs entières et non présentées sous forme d'intervalles.

(b): 

Pour verifier l'existence d'un lien entre la variable chol et la variable thalach on peut calculer le coefficient de correlation entre les deux variables. 
Pour ce faire nous pouvons utiliser la correlation de Pearson (Le coefficient de Pearson est un indice reflétant une relation linéaire entre deux variables continues. Le coefficient de corrélation varie entre -1 et +1, 0 reflétant une relation nulle entre les deux variables, une valeur négative (corrélation négative) signifiant que lorsqu'une des variable augmente, l'autre diminue ; tandis qu'une valeur positive (corrélation positive) indique que les deux variables varient ensemble dans le même sens. 
source: http://www.biostat.ulg.ac.be/pages/Site_r/corr_pearson.html
).

Processus: Le coefficient de Pearson se calcule comme suit: 
$$
r = \frac{{\sum (X_i - \bar{X}) \cdot (Y_i - \bar{Y})}}{{\sqrt{\sum (X_i - \bar{X})^2} \cdot \sqrt{\sum (Y_i - \bar{Y})^2}}}
$$


Avec :
-  X et Y   Nos deux variables entre lesquelles nous souhaitons verifier l'existence d'un lien ( X  pour chol et Y pour thalach );

- r : le coefficient de corrélation de Pearson;

Justification: L'erreur quadratique moyenne serait la fonction de perte la plus adaptée, car elle minimise considérablement les erreurs en se basant sur les différences entre les valeurs prédites et les valeurs réelles. Cela permet d'avoir une idée directe sur la performance du modèle de manière simple et rapide.
                                   
iii. Comme algorithme d'optimisation nous suggérons la methode analytique pour l’entrainement de notre modèle.

Fonctionnement: Cet algorithme consiste à trouver les valeurs des paramètres pour lesquelles la fonction perte serait minimisée. 
- Tout d'abord nous calculons les derivées partielles de notre fonction perte;
- Ensuite nous cherchons les points critiques en posant les derivées partielles egales a zero (les solutions des équations seront nos points critiques);
- Afin de determiner la nature de nos points critiques nous calculons les derivées secondes ;
- Et pour finir nous concluons avec cette logique :
 si la derivee seconde > 0 alors le point critique devient un minimum;
 si la derivee seconde < 0 alors le point critique devient un maximum;
 si la derivee seconde= 0 alors le point critique devient soit un point selle soit un point d'inflexion;

iv. Selon nous, cet algorithme peut être limité en cas de complexité du modèle, ce qui rendrait très difficile la recherche des points critiques. De plus, il pourrait prendre beaucoup de temps à résoudre dans les cas où le modèle comporte de nombreux paramètres, les calculs pourraient prendre beaucoup de temps de calcul.

Les données d'entraînement servent à apprendre au modèle afin qu'il puisse généraliser et prendre des décisions. Les données de test servent à évaluer la performance du modèle en le confrontant à des données qu'il n'a jamais vues.

Pour évaluer la qualité de notre modèle de régression, nous choisissons le coefficient de détermination R² sur nos données de test. En effet, pour mieux évaluer la qualité du modèle, il est important de le tester sur des données non étiquetées et non vues par le modèle afin d'évaluer sa capacité à généraliser.

( Ce coefficient est exprimé sur une échelle de 0 à 1, où un R² de 1 indique que le modèle de régression explique entièrement la variance de la variable dépendante, ce qui signifie une adéquation parfaite entre les valeurs prédites et les valeurs réelles.

À l’inverse, un R² de 0 suggère que le modèle ne parvient pas à expliquer la variance de la variable dépendante, indiquant un ajustement nul du modèle aux données.
source : https://blog.nalo.fr/lexique/r-carre-ou-r2/
)

Cette logique rend le coefficient facile à interpréter et à utiliser, offrant des évaluations claires et précises, pouvant même être exprimées sous forme de pourcentage.
