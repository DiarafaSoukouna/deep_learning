Pour verifier l'existence d'un lien entre la variable chol et la variable thalach on peut calculer le coefficient de correlation entre les deux variables. 
Pour ce faire nous pouvons utiliser la correlation de Pearson (Le coefficient de Pearson est un indice reflétant une relation linéaire entre deux variables continues. Le coefficient de corrélation varie entre -1 et +1, 0 reflétant une relation nulle entre les deux variables, une valeur négative (corrélation négative) signifiant que lorsqu'une des variable augmente, l'autre diminue ; tandis qu'une valeur positive (corrélation positive) indique que les deux variables varient ensemble dans le même sens. 
source: http://www.biostat.ulg.ac.be/pages/Site_r/corr_pearson.html
).

Processus: Le coefficient de Pearson se calcule comme suit: 
$$
r = \frac{{\sum (X_i - \bar{X}) \cdot (Y_i - \bar{Y})}}{{\sqrt{\sum (X_i - \bar{X})^2} \cdot \sqrt{\sum (Y_i - \bar{Y})^2}}}
$$


Avec :
-  X et Y   Nos deux variables entre lesquelles nous souhaitons verifier l'existence d'un lien ( X  pour chol et Y pour thalach );

- r : le coefficient de corrélation de Pearson;

(b):
Normalisons les données pour que les valeurs des pixels soient entre 0 et 1. Pour cela, nous allons diviser les valeurs de nos variables test_images, train_images par 255.0

(c):
Affichage de quelques images pour montrer une idée des 10 catégories d’images composant
l’ensemble de données. Pour cela nous allons utiliser la fonction imshow() de matplotlib.pyplot 

```python
def show_some_images(nrows, ncols, class_names = []):
  """
    Cette méthode affiche quelques images de notre dataset sous forme de tableau

    nrows : Le nombre de ligne
    ncols : Le nombre de colonnes
    class_names : La liste des catégories
  """
  fig, axs = plt.subplots(nrows, ncols, figsize=(8, 8), constrained_layout=True) 
  fig.suptitle('Quelques images')

  image_index = 0;
  for i in range(nrows):
      for j in range(ncols):
          axs[i, j].imshow(train_images[image_index])

          if len(class_names) > 0:
            axs[i, j].set_title(class_names[train_labels[image_index][0]])

          image_index += 1

  plt.show()
```

```python
show_some_images(6, 5)
```

6. Creation d'un modele de deep learning. Pour cela nous allons utiliser Tensorflow/Keras

```python
nb_pixel= 3072 
train_images = train_images.reshape(-1, nb_pixel)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the model
model = Sequential()

# Add the first hidden layer with 512 neurons and relu activation function
model.add(Dense(128, activation='relu', input_shape=(nb_pixel,)))

# Add the second hidden layer with 256 neurons and relu activation function
model.add(Dense(256, activation='relu'))

# Add the third hidden layer with 128 neurons and relu activation function
model.add(Dense(128, activation='relu'))

# Add the output layer with the number of classes you want to predict and softmax activation function
model.add(Dense(10, activation='softmax'))

# Build the model
model.build(input_shape=(None, nb_pixel))

```

(a):
 A l’aide de la commande summary(), évaluons et commentons la complexité du
modèle 

```python
model.summary()
```

Commentaire: Ce modèle est assez complexe en raison du grand nombre de paramètres à estimer, totalisant 393 344. Cette complexité lui permet d'apprendre des motifs complexes dans les données. Cependant, le nombre élevé de paramètres par rapport au nombre total d'exemples dans l'ensemble de données, qui est de 60 000, pourrait entraîner un risque de sur-apprentissage. 

(b): 
Nous entraînons le modèle en réservant 20% des données comme ensemble de validation. Cet ensemble de validation sera utilisé pour évaluer les performances du modèle pendant l'entraînement et analyser sa performance.

```python
# from tensorflow.keras.optimizers import SGD
# from tensorflow.keras.optimizers import Adam
model.compile(optimizer='sgd', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

# Entraînement du modèle avec vos données
model.fit(train_images, train_labels, epochs=10, validation_split=0.2)
```

(c):
L'entraînement du modèle s'est déroulé de manière satisfaisante, avec une amélioration de la précision au fil des époques, indiquant une adaptation continue du modèle et une augmentation de sa performance. De plus, les valeurs de perte ont diminué progressivement au fil des époques, ce qui suggère que le modèle a réussi à réduire les pertes au cours de l'entraînement. 
Le même schéma se répète également avec les données de validation, où la performance augmente tandis que les pertes diminuent en même temps.

(d):
Pour évaluer le pouvoir prédictif final de notre modèle, nous devons d'abord le tester sur nos données de test. Cette étape nous permettra de comprendre comment le modèle se comportera par rapport à des données qu'il n'a jamais rencontrées auparavant. En fonction des performances observées lors de ce test, nous pourrons conclure sur le pouvoir prédictif global de notre modèle. 

```python
model1= tf.keras.models.load_model('model')
loss, accuracy = model1.evaluate(test_images, test_labels)

# Afficher les résultats
print("Perte sur les données de test:", loss)
print("Précision sur les données de test:", accuracy)
```
